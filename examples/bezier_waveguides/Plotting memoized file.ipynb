```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
file = "bezier_optimal.npz"
npzfile = np.load(file)
x = npzfile["x"]
y = npzfile["y"]
z_a = npzfile["z_a"]
z_b = npzfile["z_b"]
```

```python
nx = len(x)
ny = len(y)
DIV = 5
fig, ax = plt.subplots(nrows=nx//DIV+1, ncols=1, sharex=True, sharey=True, figsize=(10, 6 * nx / DIV))
for i in range(0, nx, DIV):
    ax[i//DIV].plot(y, z_a[i,:], '.-', label='a')
    ax[i//DIV].plot(y, z_b[i,:], '.-', label='b')
    ax[i//DIV].axvline(x[i])
    ax[i//DIV].legend()
    ax[i//DIV].set_ylabel(f"alpha_0 = {x[i]}")
    ax[i//DIV].set_title("alpha_3")
plt.show()
```
